```python

%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# --- 0. Ajustes de salida (carpeta) ---
output_dir = "output_graficos"
os.makedirs(output_dir, exist_ok=True)

# --- 1. Cargar dataset ---
df = pd.read_csv("Edades1_filtrado.csv")

# --- 2. Limpieza y mapeos (según tus descripciones previas) ---
# Convertir tipos numéricos
df["BO2012A_AGE"] = pd.to_numeric(df["BO2012A_AGE"], errors="coerce")
df["BO2012A_SEX"] = pd.to_numeric(df["BO2012A_SEX"], errors="coerce")
df["BO2012A_SCHOOL"] = pd.to_numeric(df["BO2012A_SCHOOL"], errors="coerce")
df["BO2012A_EDLEV"] = pd.to_numeric(df["BO2012A_EDLEV"], errors="coerce")
df["BO2012A_RESDEPT"] = pd.to_numeric(df["BO2012A_RESDEPT"], errors="coerce")

# Mapas legibles
school_map = {1: "Publica", 2: "Privada", 3: "Convenio"}
sex_map = {1: "Femenino", 2: "Masculino"}
edlev_map = {
    3: "Inicial (Prek/Kinder)",
    4: "Basico (1-5)",
    5: "Intermedio (6-8)",
    6: "Medio (9-12)",
    7: "Primaria (ant.)",
    8: "Secundaria (ant.)",
    9: "Primaria (act.)",
    10:"Secundaria (act.)"
}
dept_map = {
    2.0: "La Paz",
    7.0: "Santa Cruz",
    3.0: "Cochabamba",
    5.0: "Potosi",
    1.0: "Chuquisaca",
    4.0: "Oruro",
    6.0: "El Alto",
    8.0: "Tarija",
    9.0: "Beni",
    10.0:"Pando"
}

df["SCHOOL_TYPE"] = df["BO2012A_SCHOOL"].map(school_map)
df["SEX_LABEL"] = df["BO2012A_SEX"].map(sex_map)
df["EDLEV_LABEL"] = df["BO2012A_EDLEV"].map(edlev_map)
df["DEPARTAMENTO"] = df["BO2012A_RESDEPT"].map(dept_map)

# Grupo etario
def edad_grupo(edad):
    if pd.isna(edad):
        return np.nan
    e = int(edad)
    if 0 <= e <= 5:
        return "Primera infancia (0-5)"
    elif 6 <= e <= 11:
        return "Niños/as (6-11)"
    elif 12 <= e <= 17:
        return "Adolescentes (12-17)"
    elif 18 <= e <= 28:
        return "Jóvenes (18-28)"
    else:
        return "Fuera rango"
df["AGE_GROUP"] = df["BO2012A_AGE"].apply(edad_grupo)

# Conteos de colegios por departamento (tus datos)
schools_by_dept = {
    "El Alto": 524,
    "La Paz": 4230,
    "Cochabamba": 3700,
    "Santa Cruz": 3400,
    "Potosi": 1500,
    "Chuquisaca": 1050,
    "Oruro": 850,
    "Tarija": 800,
    "Beni": 700,
    "Pando": 250
}

# --- 3. Tablas resumen básicas ---
# Totales por departamento
dept_group = df.groupby("DEPARTAMENTO")
summary_dept = dept_group.agg(
    total_estudiantes = ("DEPARTAMENTO", "count"),
    edad_media = ("BO2012A_AGE", "mean"),
    edad_mediana = ("BO2012A_AGE", "median"),
    edad_std = ("BO2012A_AGE", "std")
).reset_index()

# Añadir escuelas y estudiantes por escuela
summary_dept["escuelas"] = summary_dept["DEPARTAMENTO"].map(schools_by_dept)
summary_dept["estudiantes_por_colegio"] = summary_dept.apply(
    lambda r: r["total_estudiantes"]/r["escuelas"] if pd.notna(r["escuelas"]) and r["escuelas"]>0 else np.nan, axis=1
)

# Guardar resumen
summary_dept.to_csv(os.path.join(output_dir, "summary_por_departamento.csv"), index=False)

# --- 4. Pivotes y tablas auxiliares ---
pivot_dept_edlev = pd.pivot_table(df, index="DEPARTAMENTO", columns="EDLEV_LABEL", aggfunc="size", fill_value=0)
pivot_dept_agegroup = pd.pivot_table(df, index="DEPARTAMENTO", columns="AGE_GROUP", aggfunc="size", fill_value=0)
pivot_dept_school = pd.pivot_table(df, index="DEPARTAMENTO", columns="SCHOOL_TYPE", aggfunc="size", fill_value=0)
pivot_dept_sex = pd.pivot_table(df, index="DEPARTAMENTO", columns="SEX_LABEL", aggfunc="size", fill_value=0)

pivot_dept_edlev.to_csv(os.path.join(output_dir, "pivot_dept_edlev.csv"))
pivot_dept_agegroup.to_csv(os.path.join(output_dir, "pivot_dept_agegroup.csv"))
pivot_dept_school.to_csv(os.path.join(output_dir, "pivot_dept_school.csv"))
pivot_dept_sex.to_csv(os.path.join(output_dir, "pivot_dept_sex.csv"))

# --- 5. FUNCIONES AUXILIARES DE GRAFICADO ---
def guardar_y_mostrar(fig, filename):
    fig.tight_layout()
    path = os.path.join(output_dir, filename)
    fig.savefig(path, dpi=150)
    plt.show()
    plt.close(fig)

# Orden de departamentos consistente (por total)
dept_order = summary_dept.sort_values("total_estudiantes", ascending=False)["DEPARTAMENTO"].dropna().tolist()

# --- 6. GRAFICOS GENERALES (uno por uno) ---
# 6.1 Totales por departamento (barra)
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
totales = summary_dept.set_index("DEPARTAMENTO").loc[dept_order, "total_estudiantes"]
ax.bar(totales.index, totales.values)
ax.set_title("Total de estudiantes por departamento (conteo en dataset)")
ax.set_ylabel("Número de estudiantes")
ax.set_xticklabels(totales.index, rotation=45, ha='right')
guardar_y_mostrar(fig, "01_total_estudiantes_por_departamento.png")

# 6.2 Estudiantes por colegio (promedio) por departamento
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
vals = summary_dept.set_index("DEPARTAMENTO").loc[dept_order, "estudiantes_por_colegio"]
ax.bar(vals.index, vals.values)
ax.set_title("Promedio de estudiantes por unidad educativa (por departamento)")
ax.set_ylabel("Estudiantes promedio por colegio")
ax.set_xticklabels(vals.index, rotation=45, ha='right')
guardar_y_mostrar(fig, "02_estudiantes_por_colegio_por_departamento.png")

# 6.3 Distribución por EDLEV (conteo global) - barras horizontales
global_edlev = df["EDLEV_LABEL"].value_counts().sort_values(ascending=True)
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1)
ax.barh(global_edlev.index, global_edlev.values)
ax.set_title("Distribución global por nivel educativo (EDLEV)")
ax.set_xlabel("Número de estudiantes")
guardar_y_mostrar(fig, "03_distribucion_global_edlev.png")

# 6.4 Distribución por grupo etario (global)
global_agegroup = df["AGE_GROUP"].value_counts()
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(1,1,1)
ax.pie(global_agegroup.values, labels=global_agegroup.index, autopct='%1.1f%%')
ax.set_title("Distribución por grupo etario (global)")
guardar_y_mostrar(fig, "04_pie_grupo_etario_global.png")

# 6.5 Histograma de edades + líneas de media y mediana
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1)
ages = df["BO2012A_AGE"].dropna()
ax.hist(ages, bins=range(0,31), edgecolor='black')
ax.axvline(ages.mean(), linestyle='--', linewidth=1.5, label=f"Media={ages.mean():.2f}")
ax.axvline(ages.median(), linestyle=':', linewidth=1.5, label=f"Mediana={ages.median():.0f}")
ax.set_title("Histograma de edades (0-29)")
ax.set_xlabel("Edad")
ax.set_ylabel("Frecuencia")
ax.legend()
guardar_y_mostrar(fig, "05_histograma_edades.png")

# 6.6 Boxplot: edad por departamento (solo departamentos con suficientes datos)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)
# construir datos ordenados
data_for_box = [df[df["DEPARTAMENTO"]==d]["BO2012A_AGE"].dropna() for d in dept_order]
ax.boxplot(data_for_box, labels=dept_order, showfliers=False)
ax.set_title("Boxplot de edad por departamento (sin outliers mostrados)")
ax.set_ylabel("Edad")
plt.xticks(rotation=45, ha='right')
guardar_y_mostrar(fig, "06_boxplot_edad_por_depto.png")

# 6.7 Stacked bar: grupos etarios por departamento (absoluto)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)
age_cols = ["Primera infancia (0-5)", "Niños/as (6-11)", "Adolescentes (12-17)", "Jóvenes (18-28)"]
# asegurar columnas
for c in age_cols:
    if c not in pivot_dept_agegroup.columns:
        pivot_dept_agegroup[c] = 0
pivot_age = pivot_dept_agegroup.loc[dept_order, age_cols]
bottom = np.zeros(len(pivot_age))
x = np.arange(len(pivot_age))
for i, c in enumerate(age_cols):
    ax.bar(pivot_age.index, pivot_age[c].values, bottom=bottom, label=c)
    bottom += pivot_age[c].values
ax.set_title("Distribución por grupo etario por departamento (absoluto)")
ax.set_xticklabels(pivot_age.index, rotation=45, ha='right')
ax.legend()
guardar_y_mostrar(fig, "07_stacked_agegroup_por_depto.png")

# 6.8 100% Stacked bar: proporción de grupos etarios por departamento
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)
proportions = pivot_age.div(pivot_age.sum(axis=1).replace(0,np.nan), axis=0).fillna(0)
bottom = np.zeros(len(proportions))
for c in age_cols:
    ax.bar(proportions.index, proportions[c].values, bottom=bottom, label=c)
    bottom += proportions[c].values
ax.set_title("Proporción (%) de grupos etarios por departamento (normalizado)")
ax.set_xticklabels(proportions.index, rotation=45, ha='right')
ax.legend()
guardar_y_mostrar(fig, "08_100p_stacked_agegroup_por_depto.png")

# 6.9 Distribución por tipo de institución por departamento (absoluto y normalizado)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)
school_cols = ["Publica", "Privada", "Convenio"]
for c in school_cols:
    if c not in pivot_dept_school.columns:
        pivot_dept_school[c] = 0
pivot_school = pivot_dept_school.loc[dept_order, school_cols]
bottom = np.zeros(len(pivot_school))
for c in school_cols:
    ax.bar(pivot_school.index, pivot_school[c].values, bottom=bottom, label=c)
    bottom += pivot_school[c].values
ax.set_title("Distribución por tipo de institución por departamento (absoluto)")
ax.set_xticklabels(pivot_school.index, rotation=45, ha='right')
ax.legend()
guardar_y_mostrar(fig, "09_stacked_schooltype_por_depto.png")

# Normalizado (porcentaje)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)
prop_school = pivot_school.div(pivot_school.sum(axis=1).replace(0,np.nan), axis=0).fillna(0)
bottom = np.zeros(len(prop_school))
for c in school_cols:
    ax.bar(prop_school.index, prop_school[c].values, bottom=bottom, label=c)
    bottom += prop_school[c].values
ax.set_title("Proporción por tipo de institución por departamento (100%)")
ax.set_xticklabels(prop_school.index, rotation=45, ha='right')
ax.legend()
guardar_y_mostrar(fig, "10_100p_stacked_schooltype_por_depto.png")

# 6.10 Heatmap: Dept vs EDLEV counts (matriz)
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
mat = pivot_dept_edlev.loc[dept_order].fillna(0).values
im = ax.imshow(mat, aspect='auto')  # usa el colormap por defecto
ax.set_yticks(np.arange(len(pivot_dept_edlev.loc[dept_order].index)))
ax.set_yticklabels(pivot_dept_edlev.loc[dept_order].index)
ax.set_xticks(np.arange(len(pivot_dept_edlev.columns)))
ax.set_xticklabels(pivot_dept_edlev.columns, rotation=45, ha='right')
ax.set_title("Heatmap: Departamentos vs Nivel educativo (EDLEV) - conteos")
plt.colorbar(im, ax=ax)
guardar_y_mostrar(fig, "11_heatmap_depto_edlev.png")

# 6.11 Heatmap normalizado por fila (proporción EDLEV por depto)
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
mat_norm = pivot_dept_edlev.div(pivot_dept_edlev.sum(axis=1).replace(0,np.nan), axis=0).loc[dept_order].fillna(0).values
im = ax.imshow(mat_norm, aspect='auto')
ax.set_yticks(np.arange(len(pivot_dept_edlev.loc[dept_order].index)))
ax.set_yticklabels(pivot_dept_edlev.loc[dept_order].index)
ax.set_xticks(np.arange(len(pivot_dept_edlev.columns)))
ax.set_xticklabels(pivot_dept_edlev.columns, rotation=45, ha='right')
ax.set_title("Heatmap normalizado: proporción EDLEV por departamento")
plt.colorbar(im, ax=ax)
guardar_y_mostrar(fig, "12_heatmap_normalizado_depto_edlev.png")

# 6.12 Distribución por sexo global y por departamento (stacked)
global_sex = df["SEX_LABEL"].value_counts()
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(1,1,1)
ax.bar(global_sex.index, global_sex.values)
ax.set_title("Distribución por sexo (global)")
guardar_y_mostrar(fig, "13_global_by_sex.png")

# Sexo por departamento (stacked)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(1,1,1)
sex_cols = ["Femenino", "Masculino"]
for c in sex_cols:
    if c not in pivot_dept_sex.columns:
        pivot_dept_sex[c] = 0
bottom = np.zeros(len(pivot_dept_sex.loc[dept_order]))
for c in sex_cols:
    ax.bar(pivot_dept_sex.loc[dept_order].index, pivot_dept_sex.loc[dept_order][c].values, bottom=bottom, label=c)
    bottom += pivot_dept_sex.loc[dept_order][c].values
ax.set_title("Distribución por sexo por departamento (absoluto)")
ax.set_xticklabels(pivot_dept_sex.loc[dept_order].index, rotation=45, ha='right')
ax.legend()
guardar_y_mostrar(fig, "14_sex_por_depto_stacked.png")

# 6.13 Edad vs EDLEV (scatter) - con un poco de jitter para visual
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
# Transformar EDLEV a número (ya lo es en BO2012A_EDLEV)
sub = df.dropna(subset=["BO2012A_AGE", "BO2012A_EDLEV"])
# jitter
rng = np.random.RandomState(42)
x = sub["BO2012A_EDLEV"] + (rng.rand(len(sub)) - 0.5) * 0.2
y = sub["BO2012A_AGE"] + (rng.rand(len(sub)) - 0.5) * 0.4
ax.scatter(x, y, s=6)
ax.set_xlabel("EDLEV (numérico)")
ax.set_ylabel("Edad")
ax.set_title("Edad vs EDLEV (puntos con jitter)")
guardar_y_mostrar(fig, "15_edad_vs_edlev_scatter.png")

# 6.14 Cumulative distribution (edad)
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(1,1,1)
ages_sorted = np.sort(ages)
cdf = np.arange(1, len(ages_sorted)+1) / len(ages_sorted)
ax.plot(ages_sorted, cdf)
ax.set_title("Función de distribución acumulada de la edad")
ax.set_xlabel("Edad")
ax.set_ylabel("Proporción acumulada")
guardar_y_mostrar(fig, "16_cdf_edad.png")

# 6.15 Top-N: Departamentos con más/menos estudiantes (bar)
top5 = summary_dept.sort_values("total_estudiantes", ascending=False).head(5)
bot5 = summary_dept.sort_values("total_estudiantes", ascending=True).head(5)

fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(1,2,1)
ax.bar(top5["DEPARTAMENTO"], top5["total_estudiantes"])
ax.set_title("Top 5 departamentos (mas estudiantes)")
ax.set_xticklabels(top5["DEPARTAMENTO"], rotation=45, ha='right')

ax2 = fig.add_subplot(1,2,2)
ax2.bar(bot5["DEPARTAMENTO"], bot5["total_estudiantes"])
ax2.set_title("Top 5 departamentos (menos estudiantes)")
ax2.set_xticklabels(bot5["DEPARTAMENTO"], rotation=45, ha='right')
guardar_y_mostrar(fig, "17_top5_bot5_departamentos.png")

# --- 7. Estadísticas numéricas y guardado de reportes ---
# Promedios y medianas por departamento (ya en summary_dept)
summary_dept.sort_values("total_estudiantes", ascending=False).to_csv(os.path.join(output_dir, "summary_por_departamento_ordenado.csv"), index=False)

# Estadísticas por EDLEV
edlev_stats = df.groupby("EDLEV_LABEL")["BO2012A_AGE"].agg(["count","mean","median","std"]).sort_values("count", ascending=False)
edlev_stats.to_csv(os.path.join(output_dir, "edlev_stats.csv"))

# Estadísticas por tipo de colegio
school_stats = df.groupby("SCHOOL_TYPE")["BO2012A_AGE"].agg(["count","mean","median","std"]).sort_values("count", ascending=False)
school_stats.to_csv(os.path.join(output_dir, "schooltype_stats.csv"))

# 7.x Mostrar resumen en pantalla
print("\nResumen por departamento (primeras filas):")
display(summary_dept.head(20))

print("\nEstadísticas por nivel educativo (EDLEV):")
display(edlev_stats)

print("\nEstadísticas por tipo de colegio:")
display(school_stats)

print(f"\nTodas las figuras y CSVs han sido guardados en la carpeta: {output_dir}")

# FIN del script
```

```python
# -------------------- Comparacion Pandas vs PySpark (versión Windows, gráficos en pantalla) --------------------
CSV_PATH = "Edades1.csv"   # <- cambia si tu CSV está en otra ruta
REPEATS = 3

# Imports
import os, sys, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Columnas
COL_AGE = "BO2012A_AGE"
COL_SEX = "BO2012A_SEX"
COL_SCHOOL = "BO2012A_SCHOOL"

# -------------------- Helpers --------------------
def measure(func, *args, **kwargs):
    t0 = time.perf_counter()
    res = func(*args, **kwargs)
    t1 = time.perf_counter()
    return res, t1 - t0

# -------------------- Pandas pipeline --------------------
def pandas_pipeline(csv_path):
    df, read_time = measure(pd.read_csv, csv_path, low_memory=False)
    t0 = time.perf_counter()
    cols_present = [c for c in (COL_AGE, COL_SEX, COL_SCHOOL) if c in df.columns]
    df = df[cols_present].copy()
    if COL_AGE in df.columns:
        df[COL_AGE] = pd.to_numeric(df[COL_AGE], errors="coerce")
        df = df.dropna(subset=[COL_AGE])
    if COL_SEX in df.columns:
        df[COL_SEX] = df[COL_SEX].astype("object")
    if COL_SCHOOL in df.columns:
        df[COL_SCHOOL] = df[COL_SCHOOL].astype("object")
    transform_time = time.perf_counter() - t0

    t0 = time.perf_counter()
    mean_age = df[COL_AGE].mean() if COL_AGE in df.columns else None
    median_age = df[COL_AGE].median() if COL_AGE in df.columns else None
    avg_by_school = df.groupby(COL_SCHOOL)[COL_AGE].mean().sort_index() if COL_SCHOOL in df.columns and COL_AGE in df.columns else pd.Series(dtype=float)
    counts_sex = df[COL_SEX].value_counts() if COL_SEX in df.columns else pd.Series(dtype=int)
    counts_school = df[COL_SCHOOL].value_counts() if COL_SCHOOL in df.columns else pd.Series(dtype=int)
    agg_time = time.perf_counter() - t0

    # -------------------- Plots --------------------
    try:
        if COL_AGE in df.columns:
            plt.figure(figsize=(8,5))
            plt.hist(df[COL_AGE], bins=30)
            plt.title("Histograma de edades (Pandas)")
            plt.xlabel("Edad")
            plt.ylabel("Frecuencia")
            plt.show()
    except Exception as e:
        print("Pandas hist plot error:", e)

    try:
        counts_school_sel = counts_school[counts_school > 20].index
        if len(counts_school_sel) > 0 and COL_AGE in df.columns:
            df_sel = df[df[COL_SCHOOL].isin(counts_school_sel)]
            plt.figure(figsize=(10,6))
            df_sel.boxplot(column=COL_AGE, by=COL_SCHOOL, rot=45)
            plt.title("Edad por tipo de escuela (Pandas)")
            plt.suptitle("")
            plt.xlabel("Tipo de escuela")
            plt.ylabel("Edad")
            plt.show()
    except Exception as e:
        print("Pandas boxplot error:", e)

    try:
        if not counts_sex.empty:
            plt.figure(figsize=(6,4))
            counts_sex.plot(kind="bar")
            plt.title("Conteo por sexo (Pandas)")
            plt.xlabel("Sexo")
            plt.ylabel("Conteo")
            plt.show()
    except Exception as e:
        print("Pandas counts plot error:", e)

    plotting_time = time.perf_counter() - t0

    results = {
        "mean_age": float(mean_age) if mean_age is not None else None,
        "median_age": float(median_age) if median_age is not None else None,
        "avg_by_school": avg_by_school.to_dict() if not avg_by_school.empty else {},
        "counts_sex": counts_sex.to_dict() if not counts_sex.empty else {},
        "counts_school": counts_school.to_dict() if not counts_school.empty else {},
    }

    timings = {"read": read_time, "transform": transform_time, "agg": agg_time, "plot": plotting_time}
    return results, timings

# -------------------- PySpark pipeline --------------------
def pyspark_pipeline(csv_path, sample_limit=20000):
    os.environ["PYSPARK_PYTHON"] = sys.executable
    os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable
    from pyspark.sql import SparkSession
    from pyspark.sql.functions import col, mean as _mean, count as _count

    builder = SparkSession.builder \
        .appName("timing_compare") \
        .master("local[1]") \
        .config("spark.ui.showConsoleProgress", "false") \
        .config("spark.driver.bindAddress", "127.0.0.1") \
        .config("spark.driver.host", "127.0.0.1") \
        .config("spark.python.worker.reuse", "false") \
        .config("spark.network.timeout", "600s") \
        .config("spark.executor.heartbeatInterval", "60s")

    spark = builder.getOrCreate()

    t0 = time.perf_counter()
    sdf = spark.read.option("header", True).option("inferSchema", True).csv(csv_path)
    read_time = time.perf_counter() - t0

    t0 = time.perf_counter()
    cols_avail = [c for c in (COL_AGE, COL_SEX, COL_SCHOOL) if c in sdf.columns]
    sdf_sel = sdf.select(*cols_avail) if len(cols_avail) > 0 else sdf
    if COL_AGE in sdf_sel.columns:
        sdf_sel = sdf_sel.withColumn(COL_AGE, col(COL_AGE).cast("double")).na.drop(subset=[COL_AGE])
    try:
        sdf_sel.limit(10).collect()
    except:
        pass
    transform_time = time.perf_counter() - t0

    t0 = time.perf_counter()
    mean_age = None
    median_age = None
    avg_by_school = {}
    counts_sex = {}
    counts_school = {}
    try:
        from pyspark.sql.functions import expr
        if COL_AGE in sdf_sel.columns:
            mean_row = sdf_sel.agg(_mean(COL_AGE).alias("mean_age")).collect()[0]
            mean_age = mean_row["mean_age"]
            median_row = sdf_sel.selectExpr(f"percentile_approx({COL_AGE}, 0.5) as median").collect()[0]
            median_age = median_row["median"]
        if COL_SCHOOL in sdf_sel.columns and COL_AGE in sdf_sel.columns:
            rows = sdf_sel.groupBy(COL_SCHOOL).agg(_mean(COL_AGE).alias("avg_age")).orderBy(COL_SCHOOL).collect()
            avg_by_school = {r[COL_SCHOOL]: r["avg_age"] for r in rows}
        if COL_SEX in sdf_sel.columns:
            rows = sdf_sel.groupBy(COL_SEX).agg(_count("*").alias("cnt")).collect()
            counts_sex = {r[COL_SEX]: r["cnt"] for r in rows}
        if COL_SCHOOL in sdf_sel.columns:
            rows = sdf_sel.groupBy(COL_SCHOOL).agg(_count("*").alias("cnt")).collect()
            counts_school = {r[COL_SCHOOL]: r["cnt"] for r in rows}
    except:
        pass
    agg_time = time.perf_counter() - t0

    # -------------------- Plots --------------------
    t0 = time.perf_counter()
    try:
        if COL_AGE in sdf_sel.columns:
            df_sample = sdf_sel.select(COL_AGE).limit(sample_limit).toPandas()
            age_sample = df_sample[COL_AGE].to_numpy()
            if age_sample.size > 0:
                plt.figure(figsize=(8,5))
                plt.hist(age_sample, bins=30)
                plt.title("Histograma de edades (PySpark)")
                plt.xlabel("Edad")
                plt.ylabel("Frecuencia")
                plt.show()
        if counts_sex:
            s = pd.Series(counts_sex).sort_index()
            plt.figure(figsize=(6,4))
            s.plot(kind="bar")
            plt.title("Conteo por sexo (PySpark)")
            plt.xlabel("Sexo")
            plt.ylabel("Conteo")
            plt.show()
        if counts_school:
            ps_counts_school = pd.Series(counts_school).sort_values(ascending=False)
            top_schools = ps_counts_school.head(6).index.tolist()
            box_data = []
            labels = []
            for sc in top_schools:
                try:
                    df_tmp = sdf_sel.filter(col(COL_SCHOOL) == sc).select(COL_AGE).limit(5000).toPandas()
                    arr = df_tmp[COL_AGE].to_numpy()
                    if arr.size > 0:
                        box_data.append(arr)
                        labels.append(str(sc))
                except:
                    continue
            if box_data:
                plt.figure(figsize=(10,6))
                plt.boxplot(box_data, labels=labels)
                plt.title("Boxplot edad por (top) escuela (PySpark)")
                plt.xlabel("Tipo de escuela")
                plt.ylabel("Edad")
                plt.show()
    except Exception as e:
        print("Error en plotting PySpark:", e)
    plotting_time = time.perf_counter() - t0

    try:
        spark.stop()
    except:
        pass

    results = {
        "mean_age": float(mean_age) if mean_age is not None else None,
        "median_age": float(median_age) if median_age is not None else None,
        "avg_by_school": avg_by_school,
        "counts_sex": counts_sex,
        "counts_school": counts_school,
    }
    timings = {"read": read_time, "transform": transform_time, "agg": agg_time, "plot": plotting_time}
    return results, timings

# -------------------- Runner --------------------
def run_experiments(csv_path, repeats=3):
    pandas_results = []
    pyspark_results = []

    for i in range(repeats):
        print(f"\n--- Repetición {i+1}/{repeats} (Pandas) ---")
        res_p, t_p = pandas_pipeline(csv_path)
        pandas_results.append((res_p, t_p))
        print("Resultados Pandas:", res_p)
        print("Timings Pandas:", t_p)

        print(f"\n--- Repetición {i+1}/{repeats} (PySpark) ---")
        res_s, t_s = pyspark_pipeline(csv_path)
        pyspark_results.append((res_s, t_s))
        print("Resultados PySpark:", res_s)
        print("Timings PySpark:", t_s)

    return pandas_results, pyspark_results

# -------------------- Ejecutar --------------------
if __name__ == "__main__":
    print("Iniciando experimento Pandas vs PySpark (gráficos en pantalla)")
    pandas_results, pyspark_results = run_experiments(CSV_PATH, repeats=REPEATS)
```
